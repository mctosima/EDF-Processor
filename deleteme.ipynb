```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
path_1 = '/Users/martinmanullang/Developer/EDF Processor/out/signal/4_Pink_Read_alpha.csv'
path_2 = '/Users/martinmanullang/Developer/EDF Processor/out/signal/4_Pink_Read_beta.csv'
path_3 = '/Users/martinmanullang/Developer/EDF Processor/out/signal/4_Pink_Read_delta.csv'
path_4 = '/Users/martinmanullang/Developer/EDF Processor/out/signal/4_Pink_Read_gamma.csv'
path_5 = '/Users/martinmanullang/Developer/EDF Processor/out/signal/4_Pink_Read_theta.csv'
path_6 = '/Users/martinmanullang/Developer/EDF Processor/out/signal/4_Pink_Read_raw.csv'
df = pd.read_csv(path_1)
# get the first column as numpy
signals = df.iloc[:, 2].to_numpy()
```

```python
plt.figure(figsize=(20, 5))
# x-axis limit
# plt.xlim(0, 3500)
plt.plot(signals)
```

```python
# create subplot with 6 rows
fig, axs = plt.subplots(6, 1, figsize=(20, 13))

x_axs_limt = [(0, 3500), (1000, 7000), (0, 2000), (4000, 8000), (0, 2000), (0, 2000)]
title_list = ['alpha', 'beta', 'delta', 'gamma', 'theta', 'raw']

for i, path in enumerate([path_1, path_2, path_3, path_4, path_5, path_6]):
    df = pd.read_csv(path)
    signals = df.iloc[:, 2].to_numpy()
    axs[i].plot(signals)
    axs[i].set_title(title_list[i])
    axs[i].set_xlim(x_axs_limt[i])
    # x axis label = frequency
    axs[i].set_xlabel('frequency')


# tight layout
plt.tight_layout()
plt.savefig('PSD Preview.png')
plt.show()
```
